```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months=np.array(['January','February','March','April','May','June','July','August','September','October','November','December'])

temperature=np.array([15.2, 16.5, 20.3, 24.1, 29.0, 33.5, 35.0, 34.2, 30.0, 25.3, 20.1, 16.0])

Average_temp=np.mean(temperature)
print(Average_temp)
```

```python
hottest_temp=np.max(temperature)
print(hottest_temp)

coldest_temp = np.min(temperature)
print(coldest_temp)

Normalize_temp=np.std(temperature)
print(Normalize_temp)

Above_annual_avg = np.mean(temperature)
print(Above_annual_avg)
```

```python
Visualize_temp=plt.plot(months,temperature)
plt.show()
```

```python
import pandas as pd

data={
       'Order_id':[2222,2223,2224,2225,2226,],
       'Product':['Pen','Book','Pencil','Bag','Id_Card'],
       'Quantity':[ 9,5,10,1,1],
       'Price':[65,250,60,450,500]
}
df=pd.DataFrame(data)
print(df)
```

```python
df['Total_Sale'] = df['Quantity']*df['Price']
print(df)
```

```python
#Filter orders where total sale > 1000
filtered_df=df[df['Total_Sale']>1000]
print(filtered_df)
```

```python
#Find most sold product using groupby
most_sold_product=df.groupby('Product')['Quantity'].sum()
print(most_sold_product)
```

```python
#Add a Discount column based on sale amount
df['Discount']=df['Total_Sale']*0.1
print(df)
```

```python
#Export the summary to Excel
export_to_excel = df.to_excel('Summary.xlsx',index=False)
print(export_to_excel)
```
